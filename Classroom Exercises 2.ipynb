```julia
using Pkg
using Random
using PyPlot
```

```julia
ns=10
v=100
a=0.5

getC(d, n)= (1/(2*n))*sum(d.^2);
getΔ(x, wi, t)=x*wi.-t;
getdC(n, x, d)=+1/n .* transpose(x)*d;
getw(wold, alpha, dc)=wold .-alpha.*dc;

xrange=cumsum(ones(ns))/ns
noise=randn(ns)
tdata= 2 .*xrange .-6 .+randn(ns)*0.2
X=[xrange ones(ns)]
w=Array{Float64}(undef, 2, v)
w[:,1]=[1 1]
C=zeros(v)

for i in 2:v
    Delta=getΔ(X, w[:,i-1], tdata)
    C[i]=getC(Delta, ns)
    dCdw=getdC(ns, X, Delta)
    w[:,i]=getw(w[:,i-1],a,dCdw)
end
print(w[:,end])

```

```julia
plt[:plot](2:1:v, C[2:1:v])
yscale("log")
```

# 7.2.4 Linear Classifier

```julia
ns=100
m,c=1.0,-0.1
# create the data
X=2*(rand(ns,2) .-0.5)
T=1.0*(X[:,2].>(m*X[:,1] .+c))
# add some scatter (optional)
sig=0.5
X=X .+sig*randn(ns,2)
Xb=[X ones(ns)];

```

```julia
figure(figsize=(3,3))
# draw the separating line
x=collect(minimum(X[:,1]):0.1:maximum(X[:,1]))
y=m*x .+c;
# do the plots
for k=1:ns
T[k]==1 ? plot(X[k,1],X[k,2],"go") : ()
T[k]==0 ? plot(X[k,1],X[k,2],"ro") : ()
end
plot(x,y,"k-")
xlabel("x1"); ylabel("x2");
```

```julia
# functions that calculate weight update

lfun(z)=1 ./(1 .+exp.(.-z));

function MyPrediction(XX,w)
    Z=XX*w
    P=lfun(Z) # altered from linear regression
    return P
end
function MyWeightUpdate(Xb,w,TT)
    nn=length(TT)
    PP=MyPrediction(Xb,w)
    dCdw=(1/nn)*(Xb')*(PP-TT)
    return dCdw
end;
function MyCost(XX,w,TT)
    nn=length(TT)
    PP=MyPrediction(XX,w)
    # cost is altered from linear regression
    C=(-1/nn)*sum(TT.*log.(PP) .+(1 .-TT).*log.(1 .-PP))
    return C
end;

function getPercentLine(w, percent)
    return -w[1]/w[2], -(w[3]+log(1/percent -1))/w[2]
end
```

```julia
nu=10000
a=0.5
wL=zeros(3,nu)
CL=zeros(1,nu)
wL[:,1]=[1.0,-2, 3]
CL[1,1]=MyCost(Xb,wL[:,1],T)
for k=2:nu
    w=wL[:,k-1]
    dCdw=MyWeightUpdate(Xb,w,T)
    wL[:,k]=w .-a*dCdw
    CL[1,k]=MyCost(Xb,wL[:,k],T)
end
print(wL[:, end])
wL[:,end]=wL[:,end] ./ wL[3,end]

mfinal=-wL[1,end]/wL[2,end]
cfinal=-wL[3, end]/wL[2, end]

m01, c01 =getPercentLine(wL[:, end], 0.1)
m09, c09 = getPercentLine(wL[:,end], 0.9)


#pp0fit=lfun(xx*wL[1,1] .+wL[2,1]);
#ppfit=lfun(xx*wL[1,end] .+wL[2,end]);
```
